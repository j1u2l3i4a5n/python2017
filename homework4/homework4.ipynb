# Task1. (Rock-paper-scissors Version2)
## Problem Description
Design a system where two players (Player1 is user and Player2 is computer, which would choose one of ‘s’, ‘r’, ‘p’ randomly) can play the finger-guessing game. If both players throw the same shape, the game is tied and is usually immediately replayed to break the tie.
### Input
A string
### Output
A string
### Sample Output1
    Enter s, r, p for player 1: p
    Computer shows r
    You win!
### Sample Output2
    Enter s, r, p for player 1: s
    Computer shows 
    Two players tie!
    Enter s, r, p for player 1: p
    Computer shows s
    Computer wins!
    
    
### Involved Concepts
I/O operation, string, random (ex: seed)



```python
# Write your code here
import random
hand= ["r" , "s" , "p"]
def player1Hand():
    player=raw_input("Enter s, r, p for player 1:")
    if player in hand:
        return player
    else:
        print ("You can't do this.")
        player1Hand()

def gameStart():
    player1=player1Hand()
    player2=hand[int(random.random()*3)]
    print ("Computer shows " + player2)
    if player1==player2:
        print ("Two players tie!")
        gameStart()
    else:
        if (player1,player2)in[("r" , "s"),("s" , "p"),("p","r")]:
            print ("You wins!")
        else:
            print ("Computer wins!")

gameStart()
```

# Task2. (GCD, LCM)
## Problem Description
Design a system which can calculate the greatest common divisor (GCD) and least common multiple (LCM) of two integers. Please meet the requirements as follows.
### Input
Two positive integer numbers
### Output
Two strings
### Sample Output1
    Enter first value: 25
    Enter second value: 105
    GCD: 5
    LCM: 525
    
### Sample Output2
    Enter first value: 256
    Enter second value: 2988
    GCD: 4
    LCM: 191232


### Involved Concepts
I/O operation, string, arithmetic operation

```python
# Write your code here
def GCDLCM():
    num1=int(raw_input('Enter first value: '))
    num2=int(raw_input('Enter second value: '))
    GCD=1
    for i in range(min(num1,num2),0,-1):
        if num1 % i == 0 and num2 % i == 0:
            GCD = i
            break
    LCM = num1 * num2 / GCD
    print ('GCD: '+str(GCD) + '\n' + 'LCM: ' + str(LCM))
    
GCDLCM()
```
