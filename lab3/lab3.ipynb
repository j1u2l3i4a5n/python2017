```python
fileO=open('data.csv','r')
dataRaw=fileO.read()
fileO.close()
xy=dataRaw.split('\n')
xy.pop()
data=[]
for i in xy:
    data.append(map(float,i.split(',')))
print data

def mean(xory):
    a=0
    for i in data:
        a+=i[xory]
    return a/len(data)

def std(xory,u):
    a=0
    ss=0
    for i in data:
        ss+=(i[xory]-u)**2
    a=(ss/len(data))**(float(1)/2)
    return a

def corr(x,y,u1,u2,s1,s2):
    ssxy=0
    for i in data:
        ssxy+=(i[x]-u1)*(i[y]-u2)
    return ssxy/((s1*s2)*len(data))
ux=mean(0)
uy=mean(1)
sigmax=std(0,ux)
sigmay=std(1,uy)
print sigmax
rxy=corr(0,1,ux,uy,sigmax,sigmay)

print (ux,uy,sigmax,sigmay,rxy)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

datax=[]
datay=[]
for i in data:
    datax.append(i[0])
for i in data:
    datay.append(i[1])

print datax
print datay
fig=plt.figure()

plt.plot(datax,datay,'ro')
plt.axis([-30,30,-30,30])
plt.show()

fig.savefig('points_1.pdf')
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

x=[-20,20]
y=[rxy*num*sigmay/sigmax-ux/sigmax+uy for num in x]

fig=plt.figure()

plt.plot(x,y)
plt.plot(datax,datay,'ro')
plt.axis([-30,30,-30,30])
plt.show()

fig.savefig('line_1.pdf')
```

```python
dataNx=[]
dataNy=[]

for i in datax:
    dataNx.append((i-ux)/sigmax)
for idx,i in enumerate(datay):
    dataNy.append((i-uy)/sigmay)

print dataNx
print dataNy
%matplotlib inline
import matplotlib.pyplot as plt
    
fig=plt.figure()

plt.plot(dataNx,dataNy,'ro')
plt.axis([-3,3,-3,3])
plt.show()

fig.savefig('points_2.pdf')
    
```
